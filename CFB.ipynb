```python
def add_sp_to_game(game, teams_dict):
    home_team, away_team = game['home_team'], game['away_team']
    if home_team in teams_dict:
        home_team_sp = teams_dict[home_team]['S&P+']
        game['home_team_sp'] = home_team_sp
    if away_team in teams_dict:
        away_team_sp = teams_dict[away_team]['S&P+']
        game['away_team_sp'] = away_team_sp
    if home_team in teams_dict and away_team in teams_dict:
        if game['neutral_site'] == False:
            game['home_team_projected_margin'] = home_team_sp + 2.5 - away_team_sp
        else:
            game['home_team_projected_margin'] = home_team_sp - away_team_sp
    
    return game
```

```python
def sort_teams_by_conference(teams):
    conference_hash = {}
    
    for team in teams:
        conference = team['conference']
        
        if conference not in conference_hash:
            conference_hash[conference] = [team]
        else:
            conference_hash[conference].append(team)
            
    return conference_hash

detailed_conferences = sort_teams_by_conference(teams_with_conferences)
team_strings_by_conference = { conf: [team['school'] for team in arr] for conf, arr in detailed_conferences.items() }

sec_teams = team_strings_by_conference['SEC']

team_strings_by_conference
```

```python
def accumulate_games_by_team(schedule):
    team_hash = {}
    for game in schedule:
        home_team, away_team = game['home_team'], game['away_team']
        
        if home_team not in team_hash:
            team_hash[home_team] = [game]
        else:
            team_hash[home_team].append(game)
        
        if away_team not in team_hash:
            team_hash[away_team] = [game]
        else:
            team_hash[away_team].append(game)
            
    return team_hash
        
detailed_games_by_team = accumulate_games_by_team(schedule)

detailed_games_by_team
```

```python
def get_win_probabilities_for_team(team_name, detailed_games_by_team):
    res = []
    for game in detailed_games_by_team[team_name]:
        # TODO: Replace 0 with some actual logic
        proj_margin = game.get('home_team_projected_margin', 0) if game['home_team'] == team_name else -game.get('home_team_projected_margin', 0)
        rounded_margin = round(proj_margin, 1)
        win_prob = likelihoods_map[rounded_margin] if rounded_margin > 0 else 1 - likelihoods_map[abs(rounded_margin)]
        res.append({
            'home_team': game['home_team'],
            'away_team': game['away_team'],
            'win_prob': round(win_prob, 2),
            'proj_margin': round(proj_margin, 2),
        })
    return res

get_win_probabilities_for_team('Purdue', detailed_games_by_team)
```

```python
import random

def simulate_season(season):
    wins = 0
    for game in season:
        if game['win_prob'] > random.random():
            wins = wins + 1
    return wins
```

```python
def summarize(occurrences, number_of_sims):
    return { 'occurrences': occurrences, 'likelihood': occurrences / number_of_sims }

def simulate_x_seasons(num_of_sims, season):
    simulation_results = { win_ct: 0 for win_ct in range(14) }
    for _ in range(num_of_sims):
        wins = simulate_season(season)
        simulation_results[wins] += 1
    return { k: summarize(v, num_of_sims) for k, v in simulation_results.items() }
```
